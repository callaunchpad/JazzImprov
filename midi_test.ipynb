```python
import tensorflow
import pandas
import mido
```

```python
mid = mido.MidiFile('WebsterNight.mid')

def notename(notenum):
    octv = (notenum // 12)
    notes = ['C','C#','D','D#','E','F','F#','G','G#','A','A#','B']
    nt = notes[notenum % 12]
    return nt + str(octv)
            
# Eventually this will all be OOP!!!
# Running variables (these will keep track of things)
notes = []
curr_note = {} 
curr_time = 0 
for i, track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i, track.name))
    for msg in track:
        if msg.type == 'note_on': 
            
            curr_note['note'] = notename(msg.note)
            curr_note['velocity'] = msg.velocity 
            curr_time += msg.time 
            curr_note['start_time'] = curr_time
        if msg.type == 'note_off':
            curr_note['duration'] = msg.time 
            curr_time += msg.time
            notes.append(curr_note)
            curr_note = {}
print(notes)
            
```
